```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
student_names=["Jane","Mike","Susan","Andrew","Peter","Angie","Bob","Alice","Charlie","Paul"]
scores=[12,56,78,14,89,10,34,56,87,42]
```

```python
df=pd.DataFrame({"Name":student_names,"Scores":scores})
df
```

```python
plt.plot(df["Name"],df["Scores"],linestyle="-",marker='o',label='Student scores')
plt.title("Student Scores")
plt.xlabel("Student Name")
plt.ylabel("Scores")
plt.legend()
plt.show()
```
